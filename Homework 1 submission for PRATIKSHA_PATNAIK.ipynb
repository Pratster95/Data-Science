#### Gross Pay is always higher than the Annual Salary as it is the salary without any deductions or taxes. So lets find out the rows for which the Annual Salary is greater than the gross pay and then drop those rows.

#### Lets extract the weekday from the HireDate

```python
dfObj['HireDay'] = dfObj['HireDate'].apply(lambda date: pd.Timestamp(date).weekday_name)
```

```python
dfObj
```

#### Lets classify the data as per the HireDay

```python
classmap = {}
for idx, day in enumerate(dfObj['HireDay'].unique()):
    classmap[day] = float(idx)
dfObj['Label'] = dfObj['HireDay'].apply(lambda x: classmap[x])
```

### OK, that looks good.  Now lets build up a dataframe for the X (predictor) variables, and the Y (predicted) variable.

Lets consider only a subset of JobTitle column. We will consider values from Police Officer to Office Support Specialist II. 

```python
list_JobTitle=["POLICE OFFICER","POLICE OFFICER (EID)","POLICE SERGEANT","POLICE SERGEANT EID","EMERGENCY DISPATCHER","POLICE LIEUTENANT","OFFICE SUPPORT SPECIALIST III","POLICE LIEUTENANT EID","CRIMINALIST II","COMMUNITY SERVICE OFFICER","OFFICE SUPERVISOR","Police Captain","CRIME LABORATORY TECHNICIAN II","Police Major","CRIME LABORATORY TECHNICIAN I","OFFICE SUPPORT SPECIALIST II"]
```

```python
df3.JobTitle.isin(list_JobTitle)
```

```python
Filtered_df3= df3[df3.JobTitle.isin(list_JobTitle)]
```

```python
Filtered_df3.shape
```

```python
df3.shape
```

```python
Filtered_df3.head()
```

```python
classmap = {}
for idx, name in enumerate(Filtered_df3['JobTitle'].unique()):
    classmap[name] = float(idx)
Filtered_df3['Job_Title'] = Filtered_df3['JobTitle'].apply(lambda x: classmap[x])
```

```python
Filtered_df3.tail(50)
```

I am choosing HireYear and Salary_Diff=Gross Salary($)- Annual Salary($)

```python
Filtered_df3.plot.scatter(x = 'HireYear', y = 'Salary_diff', c = 'Job_Title', colormap='viridis')
```

Using the Salary_diff and HireYear we can predict the Job Title. The Amount of taxes paid by the employee and the year in which he/she was hired can help us identify the Job Title. From the above scatter plot we can see that the majority of tax payees are Police Officers and they were hired in bulk every year. Different colours show us the various Job Titles that are available in the Police Department. Different departments pay different taxes. Office Support Specialist- the one in yellow colour dots in the plot, have the least salary and hence they pay a small amount of tax.
